```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
from sqlalchemy import create_engine
```

```python
host = r'127.0.0.1' 
db = r'depressiondata' 
user = r'postgres' 
pw = r'postgres' 
port = r'5432' 
schema = r'raw_depression_data'
```

```python
db_conn = create_engine("postgresql://{}:{}@{}:{}/{}".format(user, pw, host, port, db))
```

```python
sql="select tables.table_name from information_schema.tables where (table_schema ='"+schema+"')order by 1;"
tbl_df = pd.read_sql(sql, db_conn, index_col=None)
tbl_df
```

```python
table_name = r'depression_data_100_missing'
schema = r'raw_depression_data' 

depression_raw = pd.read_sql_table(table_name, db_conn, schema)
```

```python
depression_raw.info()
```

```python
depression_raw.head(10)
```

```python
depression_raw.replace([None], np.nan, inplace=True)
```

```python
depression_raw.head(10)
```

```python
clean_depression = depression_raw.copy()
```

```python
clean_depression['Marital Status'].unique()
```

```python
clean_depression[clean_depression['Marital Status'].isna() ]
```

```python
missing_info = clean_depression[clean_depression['Marital Status'].isna()].copy()
```

```python
missing_info.head(10)
```

```python
missing_info['Marital Status'] = ['Single', 'Single']
```

```python
missing_info.head(10)
```

```python
clean_depression = clean_depression.combine_first(missing_info)
```

```python
clean_depression.head(10)
```

```python
clean_depression.info()
```

```python
table_name = r'depression_cleaned'
schema = r'cleaned_depression' 
clean_depression.to_sql(table_name, con=db_conn, if_exists='replace', index=False, schema=schema, chunksize=1000, method='multi')
```
